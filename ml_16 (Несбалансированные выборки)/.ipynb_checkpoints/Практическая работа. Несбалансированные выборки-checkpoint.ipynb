# Задание 2

Разделите датасет на обучающую и тестовую выборки в пропорции 80:20. Разделение должно быть стратифицированным по таргету Risk.

В этом и следующих заданиях используйте random_state = 1.

```python
from sklearn.model_selection import train_test_split

rnd_state = 1

train, test = train_test_split(df, random_state=rnd_state, test_size=0.2, stratify=df['Risk'])
print(train.shape, test.shape)
train
```

# Задание 3

Проведите этап очистки и подготовки данных (data preparation) и подготовьте данные к моделированию.

```python
# На стадии разведочного анализа я уже создал необходимые трансформаторы признаков для категориальных признаков.
# Поэтому я буду использовать ранее обученные преппроцессоры,
# и добавлю преоброзования RobustScaller-ом нумерических признаков

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import RobustScaler

transformer_robus = RobustScaler()

col_transformers = ColumnTransformer([("ordinal", transformer_ordinal,['Saving accounts', 'Checking account', 'Risk']),
                                     ("Ohe", transformer_ohe, ['Housing', 'Sex', 'Purpose']),
                                      ("robusted", transformer_robus, ['Age', 'Job', 'Credit amount', 'Duration'])
                                     ])

col_transformers.fit_transform(train)

train_clean = pd.DataFrame(columns=list(col_transformers.get_feature_names_out()), data=col_transformers.fit_transform(train))
train_clean
```

```python
train_clean.columns
```

# Задание 4

Сбалансируйте обучающую выборку методом SMOTE и обучите модель RandomForestClassifier. Подберите оптимальные гиперпараметры с помощью GridSearch. Посчитайте метрику ROC-AUC на тестовой выборке.

```python
from imblearn.over_sampling import SMOTE

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

X, y = train_clean.drop(columns=['ordinal__Risk']), train_clean['ordinal__Risk']
print("befor upsampling", X.shape, y.shape)
os_smote = SMOTE(random_state=rnd_state)

X_train, y_train = os_smote.fit_resample(X, y)
print("after resampling", X_train.shape, y_train.shape)

p_grid = {'n_estimators': [50, 100, 200],
         'min_samples_split': [2, 5, 10],
         'max_features': ('sqrt', 'log2', None)}

model_forestcls = RandomForestClassifier(random_state=rnd_state)

grid_searchCV = GridSearchCV(estimator=model_forestcls, param_grid=p_grid, scoring = 'roc_auc', n_jobs=-1)

grid_searchCV.fit(X_train, y_train)
print(grid_searchCV.best_score_)
print(grid_searchCV.best_estimator_)
print(grid_searchCV.best_params_)
```
