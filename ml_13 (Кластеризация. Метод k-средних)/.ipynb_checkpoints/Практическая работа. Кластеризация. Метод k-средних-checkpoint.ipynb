## Задача

Пусть имеется набор данных `Mall_Customers.csv` о покупателях супермаркета:

* CustomerID — идентификатор покупателя;
* Genre — пол;
* Age — возраст;
* Annual Income (k$) — годовой доход покупателя в тыс. долл.;
* Spending Score (1–100) — рейтинг трат покупателя.

Решите задачу кластеризации покупателей, чтобы в дальнейшем планировать маркетинговые акции для каждой группы.

### Задание 1

Загрузите датасет и выполните разведочный анализ данных. Сделайте выводы о найденных закономерностях, особенностях и других обнаруженных свойствах данных.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Mall_Customers.csv")
print(df.head())

# Построим pairplot
sns.pairplot(df.drop(columns=['CustomerID']), hue='Genre', diag_kind='hist', palette='Set1')

# Отобразим график
plt.show()
```

```python
print(df['Genre'].value_counts())
print("Хочу преоброзовать данные о Гендере в вид 1-0, где 1 будет самый популярный гендер Female")
df['Genre_is_Female'] = df['Genre'].apply(lambda x: 1 if x=='Female' else 0)
df[['Genre_is_Female', 'Genre']]
```

```python
# Удалим старую колонку о поле Genre и CustomerID айди не думаю что должен влиять на результат кластеризации.
df.drop(columns=['Genre', 'CustomerID'], inplace=True)
print(df.describe())
df
```

```python

from sklearn.preprocessing import StandardScaler

std_scl = StandardScaler()
narr_std = std_scl.fit_transform(df)
narr_std

df_std = pd.DataFrame(data=narr_std, columns=std_scl.feature_names_in_)
df_std
```

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df_std.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12); 
```

```python

sns.scatterplot(data=df_std, x='Spending Score (1-100)', y='Annual Income (k$)');
# корреляции между Spending Score и Annual Income не наблюдается но
# По графикам видно что есть возможность разделить клиентов на 4-5 групп на основе данных Spending Score и Annual Income
```

```python
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE

k_means = KMeans(n_clusters=5)
preclasters_kmeans = k_means.fit_predict(df_std)
# preclasters
print(df_std.shape, preclasters_kmeans.shape)
df_new = df_std.copy()

df_new['preclaster_kmean'] = preclasters_kmeans
```

```python
sns.scatterplot(data=df_new, x='Spending Score (1-100)', y='Annual Income (k$)', hue='preclaster_kmean', palette="deep");
```

```python
# Инициализируем 3D-ось
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
# Добавляем точки на график
ax.scatter(df_std['Annual Income (k$)'],df_std['Spending Score (1-100)'], df_std['Age'],  marker='o')

# Названия осей
ax.set_xlabel('Age Label')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')

# Отобразим график
plt.show()
```

### Задание 4

Постройте модель кластеризации с помощью метода k-средних. Определите оптимальное количество кластеров с помощью метода локтя.

```python
c_distance_metric_WCSS = []
for k in range(2,15):
    k_means = KMeans(random_state=42, n_clusters=k)
    transform_k_means = k_means.fit_transform(df_new)
    c_distance_metric_WCSS.append(k_means.inertia_)
    
plt.plot(range(2, 15), c_distance_metric_WCSS);
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

print("Проверю все резкие переходы 4,6,10. Но оставлю как ответ 4. Как мне кажется 6-10 класстеров для не обьемного по фитчам датафрейма будет слишком усложнять интерпритацию.")
```

При работе с данными важно не только качество моделей машинного обучения, но и интерпретация полученных результатов и представление их в виде, понятном пользователям и заказчикам.

После построения модели кластеризации вы разбили покупателей на группы. Теперь важно интерпретировать эти группы, то есть создать профиль группы. 

В качестве профиля группы часто используют её типичного представителя. В кластерах типичным представителем можно назвать центроиду кластера. В sklearn получить список центроид можно с помощью атрибута `cluster_centers_` у объекта класса KMeans (этот атрибут возвращает матрицу с размерностями (количество кластеров, количество признаков), подробнее об этом [в материале](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)). 

Получите центроиды кластеров, преобразуйте их к датафрейму, опишите типичного представителя каждой группы и чем группы отличаются друг от друга. При описании и сравнении используйте методы визуализации.

```python
print("результатом fit_transform является растояние до каждой центроиды класстера. Значит поместив результат алгоритма в датафрейм с данными я могу просто достать максимально приблежонную запись к центроиде, взяв просто минимум по каждой колонке fit_transform")
k_means.cluster_centers_
```

```python

df_new[k_means.get_feature_names_out()] = transform_k_means
centroids = k_means.cluster_centers_

# Достанем представителей класса выбрав ближайшего к каждой центроиде.
typical_clas_users = df_new.loc[(df_new['kmeans0']==df_new['kmeans0'].min()) | 
    (df_new['kmeans1']==df_new['kmeans1'].min()) |
    (df_new['kmeans2']==df_new['kmeans2'].min()) |
    (df_new['kmeans3']==df_new['kmeans3'].min()), : ]

# Преоброзуем масштабированные признаки в читаемый вид
inverse_scaller = std_scl.inverse_transform(typical_clas_users[std_scl.feature_names_in_])
typical_clas_users.loc[ : , std_scl.feature_names_in_] = inverse_scaller

# Добавим колонку с результатом классификации
typical_clas_users['class'] = k_means.labels_[typical_clas_users.index]
typical_clas_users
```
